# Visu KPIs DPE

### Load KPI file

On choisit le fichier correspondant à l'année que l'on souhaite visualiser

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/6-kpis/td001_dpe_kpi.csv", dtype={'code_departement': str})
```

### Load GeoJSON départements

```python
import geopandas as gpd
```

```python
departements_geo = gpd.read_file('../external/departements-version-simplifiee.geojson')
```

```python
departements_geo
```

## Visu

```python
import folium
```

```python
m = folium.Map(location=[46.759281,2.401860], tiles="OpenStreetMap", zoom_start=6)
```

```python
df['code'] = df['code_departement'].astype(str)
```

```python
m.choropleth(
    geo_data=departements_geo,
    name="Statistiques DPE",
    data=df,
    columns=['code', 'percent_eco_abc'],
    key_on='feature.properties.code',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Statistiques Départements"
    )
```

```python
df
```

```python
for index, row in df.iterrows():
    test = folium.Html('<b>Bilan DPE Consommation Energie:</b><br/><br/>Nombre : '+str(row['Nb'])+'<br/><br/>Classe A : '+str(round(row['stat_eco_a'],2))+'%<br/>Classe B : '+str(round(row['stat_eco_b'],2))+'%<br/>Classe C : '+str(round(row['stat_eco_c'],2))+'%<br/>Classe D : '+str(round(row['stat_eco_d'],2))+'%<br/>Classe E : '+str(round(row['stat_eco_e'],2))+'%<br/>Classe F : '+str(round(row['stat_eco_f'],2))+'%<br/>Classe G : '+str(round(row['stat_eco_g'],2))+'%<br/>Classe N : '+str(round(row['stat_eco_n'],2))+'%', script=True)
    popup = folium.Popup(test, max_width=200)
    folium.Marker([row['LatDD'],row['LonDD']], popup= popup, tooltip=row['nom_departement'], icon=folium.Icon()).add_to(m)
```

# Carte ges par département

```python
m2 = folium.Map(location=[46.759281,2.401860], tiles="OpenStreetMap", zoom_start=6)
m2.choropleth(
    geo_data=departements_geo,
    name="Statistiques DPE",
    data=df,
    columns=['code', 'percent_ges_abc'],
    key_on='feature.properties.code',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Statistiques Départements"
    )
```
